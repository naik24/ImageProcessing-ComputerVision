## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from sklearn.cluster import KMeans
```

## Image Segmentation using KMeans Clustering

```python
# loading the original image
butterfly = cv2.imread('butterfly.png')
Z = butterfly.reshape((-1,3))

# convert to np.float32
Z = np.float32(Z)

# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now convert back into uint8, and make original image
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((butterfly.shape))

# displaying the results
plt.subplot(1, 2, 1); plt.axis('off'); plt.title("Original Image"); plt.imshow(butterfly)
plt.subplot(1, 2, 2); plt.axis('off'); plt.title("Original Image"); plt.imshow(res2)
```
