## Importing Libraries


```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter, ImageOps
```

## Box Filter Kernels

```python
# loading the image
test_image = cv2.imread('test_pattern.tif')

# defining box filters
box_kernel_3 = np.ones((3, 3), np.float32) / 9
box_kernel_11 = np.ones((11, 11), np.float32) / 121
box_kernel_21 = np.ones((21, 21), np.float32) / 441

# performing convolution operation on the image
filtered_3 = cv2.filter2D(test_image, -1, box_kernel_3)
filtered_11 = cv2.filter2D(test_image, -1, box_kernel_11)
filtered_21 = cv2.filter2D(test_image, -1, box_kernel_21)

# displaying the results
plt.subplot(2, 2, 1)
plt.axis('off')
plt.title('Test Image')
plt.imshow(test_image, cmap = 'gray')

plt.subplot(2, 2, 2)
plt.axis('off')
plt.title('3x3 Kernel')
plt.imshow(filtered_3, cmap = 'gray')

plt.subplot(2, 2, 3)
plt.axis('off')
plt.title('11x11 Kernel')
plt.imshow(filtered_11, cmap = 'gray')

plt.subplot(2, 2, 4)
plt.axis('off')
plt.title('21x21 Kernel')
plt.imshow(filtered_21, cmap = 'gray')
```

## Gaussian Filter Kernels

```python
# loading the image
test_image = cv2.imread('test_pattern.tif')

# defining gaussian kernels
gaussian_21 = cv2.GaussianBlur(test_image, ksize = (21, 21), sigmaX = 3.5)
gaussian_43 = cv2.GaussianBlur(test_image, ksize = (43, 43), sigmaX = 7)

# displaying the results
plt.subplot(1, 3, 1)
plt.axis('off')
plt.title('Test Image')
plt.imshow(test_image)

plt.subplot(1, 3, 2)
plt.axis('off')
plt.title('21x21 Gauss Kernel')
plt.imshow(gaussian_21)

plt.subplot(1, 3, 3)
plt.axis('off')
plt.title('43x43 Gauss Kernel')
plt.imshow(gaussian_43)
```

## Order Statistic (NonLinear) Filters

```python
# loading the image
circuit_board = Image.open('circuit_board.tif')
circuit_board = ImageOps.grayscale(circuit_board)

# max filtering
max_filtered = circuit_board.filter(ImageFilter.MaxFilter(3))

# min filtering
min_filtered = circuit_board.filter(ImageFilter.MinFilter(3))

# median filtering
median_filtered = circuit_board.filter(ImageFilter.MedianFilter(3))

#displaying results
plt.subplot(2, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(circuit_board, cmap = 'gray')

plt.subplot(2, 2, 2)
plt.axis('off')
plt.title('Max Filtering')
plt.imshow(max_filtered, cmap = 'gray')

plt.subplot(2, 2, 3)
plt.axis('off')
plt.title('Min Fltered')
plt.imshow(min_filtered, cmap = 'gray')

plt.subplot(2, 2, 4)
plt.axis('off')
plt.title('Median Filtered')
plt.imshow(median_filtered, cmap = 'gray')
```
