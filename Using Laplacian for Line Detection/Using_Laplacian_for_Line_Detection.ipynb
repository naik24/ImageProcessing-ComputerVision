## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Using Laplacian for Line Detection

```python
# loading the original image
wirebond = cv2.imread('wirebond.tif')
wirebond = cv2.cvtColor(wirebond, cv2.COLOR_BGR2GRAY)
ret, wirebond = cv2.threshold(wirebond, 128, 255, cv2.THRESH_BINARY)

# defining the laplacian kernel
laplacian_kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

# applying the kernel on the image
wirebond_laplacian = cv2.filter2D(wirebond, -1, laplacian_kernel)

# displaying the results
plt.subplot(1, 2, 1); plt.axis('off'); plt.imshow(wirebond, 'gray'); plt.title('Original Image')
plt.subplot(1, 2, 2); plt.axis('off'); plt.imshow(wirebond_laplacian, 'gray'); plt.title('Laplacian')
```
