# Computer Vision

In formulating and solving computer vision problems, it's useful to draw inspiration from four high-level approaches:

- Scientific: build detailed models of the image formation process and develop mathematical techniques to invert these in order to recover the quantities of interest

- Statistical: use probabilistic models to quantify the prior likelihood of unknowns and the noisy measurement processes that produce the input images, then infer the best possible estimates of your desired quantities and analyze their resulting uncertainties

- Engineering: develop techniques that are simple to describe and implement but that are also known to work well in practice. Test these techniques to understand their limitation and failure models, as well as their expected computational costs

- Data-driven: collect a representative set of test data and use these data to either tune or learn your model parameters, or at least to validate and quantify its performance

A strong emphasis should be on testing algorithms. The best way to validate our validate your algorithms is to use three part strategy.

First, test your algorithm on clean synthetic data, for which the exact results are known. Second, add noise to the data and evaluate how the performance degrades as a function of noise level. Finally, test the algorithms on real-world data, preferably drawn from a wide variety of sources, such as photos found on the web

### Notation
- Vectors: $\textbf{v}$
- Matrices: $\textbf{M}$
- Scalars: $(T, s)$

Unless otherwise noted, vectors operate as column vectors, i.e., they post-multiply matrices, $Mv$, although they are sometimes written as comma-separated parenthesized lists x = $(x, y)$.

Some commonly used matrices are $\textbf{R}$ for rotations, $\textbf{K}$ for calibration matrices, and $\textbf{I}$ for the identity matrix. 

Homogenerous coordinates are denoted with a tilde over the vector: $\tilde{x} = (\tilde{x}, \tilde{y}, \tilde{w}) = \tilde{w}(x, y, 1) = \tilde{w}\bar{x}$
