## 1. Introduction

Interest in Digital Image Processing methods stems from two principal application areas:
- improvement of pictorial information for human interpretation
- processing of image data for tasks such as storage, transmission, and extraction of pictorial information

#### 1.1 Digital Image Processing

An image may be defined as a two-dimensional function, $f(x, y)$, where $x$ and $y$ are spatial coordinates, and the amplitude of $f$ at any pair of coordinates $(x, y)$ is called the intensity or gray level of the image at that point. A digital image is composed of a finite number of elements, each of which has a particular location and value. These elements are called *pixels*.

#### 1.2 Fundamental Steps in Digital Image Processing
It is helpful to divide steps in Image Processing into two categories:
1. Methods whose input and output are images
    - Image Acquisition
    - Image Filtering and Enhancement
    - Image Restoration
    - Wavelets and other image transforms
    - Compression and watermarking
    - Morphological Operations
2. Methods whose inputs may be images, but whose outputs are attibutes extracted from images
    - Segmentation
    - Feature extraction
    - Image Pattern Classification

## 2. Digital Image Fundamentals

### 2.1 Image Formation Model

An image is denoted by a 2-dimensional function $f(x,y)$. This function is characterized by two components:
1. $i(x,y)$: Amount of source illumination incident on the scene being viewed (*illumination*), and
2. $r(x,y)$: Amount of illumination reflected by objects in the scene (*reflectance*)

These two components combine to form $f(x,y)$:

$\boxed{f(x,y) = i(x,y)r(x,y)}$ **(Eq 2-1)**

where $0 \le i(x,y) < \infty$ and $0 \le r(x,y) \le 1$ **(Eq 2-2)**

Let the intensity (gray level) of a monochrome image at any coordinates $(x,y)$ be

$\boxed{l = f(x,y)}$ **(Eq 2-3)**

From equation (2-2) it is evident that $l$ lies in the range

$\boxed{L_{min} \le l \le L_{max}}$ **(Eq 2-4)**

The interval $[L_{min}, L_{max}]$ is called the intensity scale.

### 2.2 Image Sampling and Quantization

Our objective is to generate digital images from sensed data. The output of most sensors is a continuous voltage waveform whose amplitude and spatial behaviour are related to the physical phenomenon being sensed. To create a digital image, we need to convert the continuous sensed data into a digital format. This is done by two processes: sampling and quantization



The above image (Source: DIP Gonzalez and Woods 4e) is what we want to convert to digital form. An image may be continuous with respect to the x and y-coordinates, and also in amplitude. To digitize it, we have to sample the function in both coordinates and also in amplitude. *Digitizing the coordinate values is called sampling. Digitizing the amplitude values is called quantization.*

### 2.3 Representing Digital Images

Let $f(s, t)$ represent a continuous image function of two continuous variables, $s$ and $t$. We convert this function into digital image using sampling and quantization. Suppose that we sample the continuous image into a digital image $f(x, y)$, containing $M$ rows and $N$ columns, where $(x, y)$ are discrete coordinates. We define the origin of an image at the top left corner. 

The center of an $M x N$ image with origin at (0,0) and range $(M-1, N-1)$ is obtained by dividing $M$ and $N$ by 2 and rounding down to the nearest integer. 

```python
import math

M = 1023
N = 1024

center_M = math.floor(M/2)
center_N = math.floor(N/2)

print(f'Center: ({center_M}, {center_N})')
```

**Formal Definition of Sampling and Quantization**

Let $Z$ and $R$ denote set of integers and set of real numbers, respectively. The sampling process may be viewed as partitioning the $xy$-plane into a grid, with the coordinates of the center of each cell in the grid being a pair of elements from the Cartesian product $Z^{2}$. Hence, $f(x, y)$ is a digital image if $(x, y)$ are integers from $Z^{2}$ and $f$ is a function that assigns an intensity value (a real number from the set of real numbers $R$) to each distinct pair of coordinates. 

Image digitization requires that decisions be made regarding the values of $M, N$ and for the number, $L$, of discrete intensity levels. There are no restrictions on $M$ and $N$, other that they have to be positive integers. However, digital storage and quantizing hardware considerations usually lead to the number of intensity levels, being an integer power of two; that is 

$L = 2^{k}$ **(Eq 2-5)**

where $k$ is an integer. 

The range of values spanned by the gray scale is referred to as *dynamic range*. Dynamic range of an imaging system is the **ratio of maximum measurable intensity to the minimum detectable intensity level in the system.**

The number $b$, of bits required to store a digital image is 

$b = M \times N \times k$ **(Eq 2-6)**

### **2.4 Spatial and Intensity Resolution**
*Spatial resolution* is a measure of the smallest discernible detail in an image. Quantitatively, it can be stated as *line pairs per unit distance*, *dots per unit distance*, etc.

*Intensity resolution* refers to the smallest discernible change in the intensity level. Based on hardware considerations, the number of intensity levels is an integer power of two. Most common number is 8 bits, 16 bits. 

### **2.5 Image Interpolation**
Interpolation is used in tasks such as zooming, shrinking, rotating, and geometrically correcting digital images. Interpolation is the process of using known data to estimate values at unknown locations. 

- **Near Neighbor Interpolation**: Suppose that we have an image of size 500x500 pixel which has to be enlarged 1.5 times to 750x750 pixels. A simple way to visualize zooming is to create an imaginary 750x750 grid with the same pixel spacing as the original image, then shrink it so that it exactly overlays the original image. To assign an intensity value to any point in the overlay, we look for its closest pixel in the underlaying original image and assign the intensity of that pixel to the new pixel in the 750x750 grid. When intensities have been assigned to all the points in the overlay grid, we expand it back to the specified size to obtain the resized image. This method is simple but it has the tendency to produce undersirable artifacts, such as severe distortion of straight edges.

- **Bilinear Interpolation**: In this method, we use the four nearest neighbors to estimate the intensity of the given location. Let $(x,y)$ denote the coordinates of the location to which we want to assign an intensity value, and let $v(x,y)$ denote that intensity value. This intensity value is obtained using the equation $v(x, y) = ax + by + cxy + d$ where the four coefficients are determined from the four equations in four unknowns that can be written using the four nearest neighbors of point $(x, y)$

- **Bicubic Interpolation**: Bicubic interpolation involves sixteen nearest neighbors of a point. The equation of the intensity value assigned to point $(x, y)$ is given by $v(x, y) = \sum_{i=0}^{3} \sum_{j=0}^{3} a_{ij}x^iy^j$

An important illustration of interpolation is given here: https://x-engineer.org/bilinear-interpolation/

```python
# interpolation
import cv2
import matplotlib.pyplot as plt

#input image
image = cv2.imread('images\map.tif')

# nearest neighbor interpolation
'''
Here, after resizing the image, we get a numpy array. To display the numpy array as an image, we
can either use the pillow library to convert it from array to image or we can save it in an image 
format like JPEG and then use the imshow() function in OpenCV. 

In case of plotting the images using the matplotlib library, there's no need to convert the numpy array
into Image.
'''
nearest_neighbor_interpolation = cv2.resize(image, (1750, 1750), interpolation = cv2.INTER_NEAREST)

# bilinear interpolation
bilinear_interpolation = cv2.resize(image, (1750,1750), interpolation = cv2.INTER_LINEAR)

# bicubic interpolation
bicubic_interpolation = cv2.resize(image, (1750,1750), interpolation = cv2.INTER_CUBIC)

# displaying all the images
images = [image, nearest_neighbor_interpolation, bilinear_interpolation, bicubic_interpolation]
title = [f"Original Image {image.shape}", f"Nearest Neighbor Interpolation {nearest_neighbor_interpolation.shape}", 
         f"Bilinear Interpolation{bilinear_interpolation.shape}", f"Bicubic Interpolation{bicubic_interpolation.shape}"]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(title[i])
    plt.imshow(images[i])

plt.show()
```

### 2.6 Basic Relationships between Pixels

**2.6.1 Neighbors of a Pixel**
A pixel $p$ at coordinates $(x, y)$ has two horizontal and two vertical neighbors with coordinates $(x+1, y), (x-1, y), (x, y+1), (x, y-1)$. This set of pixels is called the 4 neighbors of $p$, is denoted by $N_{4}(p)$.

The four diagonal neighbors of $p$ have coordinates $(x+1, y+1), (x+1, y-1), (x-1, y+1), (x-1, y-1)$ are denoted $N_{D}(p)$. These neighbors together with the 4-neighbors, are called the 8-neighbors of $p$, denoted by $N_{8}(p)$.

The set of image locations of the neighbors of a point $p$ is called the neighborhood of $p$. The neighborhood is said to be closed if it contains $p$. Otherwise, the neighborhood is said to be open.

**2.6.2 Adjacency, Connectivity, Regions, and Boundaries**
Let $V$ be the set of intensity values used to define adjacency. In a binary image, $V = {1}$ if we are referring to adjaceny of pixels with value 1. For grayscale images, $V$ can contain any values between 0 and 255.

1. 4-adjacency: Two pixels $p$ and $q$ with values from $V$ are 4-adjacent if $q$ is in the set $N_{4}(p)$
2. 8-adjacency: Two pixels $p$ and $q$ with values from $V$ are 8-adjacent if $q$ is in the set $N_{8}(p)$
3. m-adjacency: Two pixels $p$ and $q$ with values from $V$ are 8-adjacent if
    - $q$ is in $N_{4}(p)$ or
    - $q$ is in $N_{D}(p)$ and the set $N_{4}(p) \cap N_{4}(q)$ has no pixels whose values are from $V$
    
**2.6.3 Distance Measures**
For pixels $p, q, s$, with coordinates $(x, y), (u, v), (w, z)$, respectively, $D$ is a distance function or metric if
1. $D(p, q) \ge 0$
2. $D(p, q) = D(q, p)$
3. $D(p, s) \le D(p, q) + D(q, s)$

The ***Euclidean Distance*** between $p$ and $q$ is defined as: $ \boxed{D_{e}(p, q) = [(x-u)^{2} + (y-v)^{2}]}$ **(Eq 2-7)**

The $D_{4}$ distance (city-block distance) is defined as: $\boxed{D_{4}(p, q) = |{x-u}| + |{y-u}|}$ **(Eq 2-8)**

The $D_{8}$ distance (city-block distance) is defined as: $\boxed{D_{8}(p, q) = max(|{x-u}|, |{y-u}|)}$ **(Eq 2-9)**

```python
# distance measures
import math

# defining point p(x, y) and q(u, v)
point_px, point_py = 10, 11

point_qu, point_qv = 15, 20

# Euclidean distance
d_euclidean = math.sqrt((point_px - point_qu)**2 + (point_py - point_qv)**2)
print("Euclidean Distance: ", d_euclidean)

# d4 distance
d4_distance = abs(point_px - point_qu) + abs(point_py - point_qv)
print("D4 distance: ", d4_distance)

# d8 distance
d8_distance = max(abs(point_px - point_qu), abs(point_py - point_qv))
print("D8 distance: ", d8_distance)
```

### 2.7 Introduction to Basic Mathematical Operations

**2.7.1 Elementwise versus Matrix Operations**

An *elementwise operation* involving one or more images is carried out on a pixel-by-pixel basis. On the other hand, *matrix product* of images is formed using rules of matrix multiplication

**2.7.2 Linear versus Nonlinear Operations**
Consider a general operator $\mathcal{H}$ that produces an output image, $g(x, y)$ from a given imput image, 

$\boxed{f(x, y): \mathcal{H}[f(x, y)] = g(x, y)}$ **(Eq 2-10)**

Given two arbitrary constants, a and b, and two arbitrary images $f_{1}(x, y)$ and $f_{2}(x, y)$, $\mathcal{H}$ is said to be linear operator if $\boxed{\mathcal{H}[af_{1}(x, y) + bf_{2}(x, y)] = a\mathcal{H}[f_{1}(x, y)] + b\mathcal{H}[f_2(x, y)] = ag_{1}(x, y) + bg_{2}(x, y)}$ **(Eq 2-11)**

**2.7.3 Arithmetic Operations**
Arithmetic operations between two images $f(x, y)$ and $g(x, y)$ are denoted as:

$s(x, y) = f(x, y) + g(x, y)$
$d(x, y) = f(x, y) - g(x, y)$
$p(x, y) = f(x, y) \times g(x, y)$
$v(x, y) = f(x, y) \div g(x,y)$

These are elementwise operations.

**NOTE**: Most images are displayed using 8 bits. Thus, we expect image values to be in the range 0 to 255. When images are saved in a standard image format, conversion to this range is automatic. When image values exceed the allowed range, clipping or scaling becomes necessary. 

Given a digital image $g$ resulting from one or more arithmetic operations, an approach guaranteeing that the full range of values is captured into a fixed number of bits is as follows:

1. First, we perform the operation:
    - $g_{m} = g - min(g)$ **(Eq 2-12)**. This creates an image whose minimum value is 0
2. Then we perform 
    - $g_{s} = K[g_{m} / max(g_{m})]$ **(Eq 2-13)** which creates a scaled image $g_{s}$, whose values are in the range $[0, K]$ 

### 2.8 Spatial Operations
Spatial operations are classified into three broad categories:
1. single-pixel operations
2. neighborhood operations
3. geometric spatial transformations

**2.8.1 Single Pixel Operations**
The simplest operation we perform on a digital image is to alter the intensity of its pixels individually using a tranformation function $T$:

$s = T(z)$ **(Eq 2-14)**

where $z$ is the intensity of the pixel in the original image. 

**2.8.2 Neighborhood Operations**
Let $S_{xy}$ denote the set of coordinates of a neighborhood centered on an arbitrary point $(x, y)$ in an image $f$. Neighborhood processing generates a corresponding pixel at the same coordinates in an outpu image, such that the value of that pixel is determined by a specified operation on the neighborhood of pixels in the input image with coordinates in the set $S_{xy}$.

**2.8.3 Geometric Transformations**
Geometric transormations are used to modify the spatial arrangement of pixels in an image. These transformations are called *rubber-sheet transformations* because they may be viewed as analogous to printing an image on a rubber sheet, then stretching or shrinking the sheet according to a predefined set of rules. 

Geometric transformations of digital images consist of two operations:
1. Spatial transformation of coordinates
2. Intensity interpolation that assigns intensity values to the spatially transformed pixels

Spatial transformation of coordinates includes *affine transformations*, which includes scaling, translation, rotation, and shearing. Affine transformation can be expressed using a single 3x3 matrix in the following form:

$\begin{bmatrix} x'\\ y'\\ z'\end{bmatrix} = $A$ \begin{bmatrix} x\\ y\\ z\end{bmatrix}$

This transformation can scale, rotate, translate, or sheer an image, depending on the values chosen for the elements of the matrxi A. 

A significant advantage of being able to perform all transformations using the unified representations is that it provides the framework for concatenating a sequence of operations. For example, if we want to resize and image, rotate it, and move the result to some location, we simply form a 3x3 matrix equal to the product of the scaling, rotation, and translation matrices.

**2.8.4 Image Registration**
In image registration, we have available an input image and a reference image. The objective is to transform the input image geometrically to produce an output image that is aligned with the reference image. 

Examples of image registration includes aligning two or more images taken at approximately the same time, but using different imaging systems. Or, the images were taken at different times using the same instruments. In such cases, combining images or performing quantitative analysis and comparisons between them requires compensating for geometric distortions caused by differences in viewing angle, distance, orientation, and other factors. 

One of the principal approaches for solving this problem is to use *tie points* (also called control points). These are corresponding points whose location are known precisely in the input and reference images. 

The problem of estimating the transformation function is one of modeling. Suppose that we have a set of four tie points, each in an input and a reference image. A simple model based on a bilinear approximation is givem by 

$x = c_{1}v + c_{2}w + c_{3}vw + c_{4} and y = c_{5}v + c_{6}w + c_{7}vw + c_{8}$ **(Eq 2-15)**

$(v,w) and (x,y)$ are the coordinates of tie points in the input and reference image respectively. If we have four pairs of corresponding tie points in both the images, we can write 8 equations using the above equations and use them to solve for the eight unknown coefficients.

In situations where four tie points are insufficient to obtain satisfactory registration, an approach used frequently is to select a larger number of tie points and then treat the quadrilaterals formed by groups of four tie points as subimages. The subimages are processed as above. Then we move to another set of four points and repeat the procedure.

### 2.9 Vector and Matrix Operations
Multispectral image processing is an area in which vector and matrix operations are used routinely. 

The *inner product* of two n-dimensional column vectors **a** and **b** is defined as:

$a \cdot b = a^{T}b$ **(Eq 2-16)**

The Euclidean vector norm, denoted by ||z|| is defined as the square root of the inner product:

$ ||z|| = (z^{T}z)^{1/2}$ **(Eq 2-17)**

We recognize this expression as length of **z**

### 2.10 Image Transforms

In some cases, image processing tasks are best formulated by transforming the input images, carrying the specified task in a *transform domain*. 

An important class of 2D linear transforms, denoted $T(u, v)$, can be expressed in the general form

$T(u, v) = \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x,y)r(x, y, u, v)$ **(Eq 2-18)**

where $f(x, y)$ is an input image, $r(x, y, u, v)$ is called a forward transformation kernel and Eq 2-18 is evaluated for u = 0, 1, 2, ..., M-1 and v = 0, 1, 2, ... , N-1. $T(u, v)$ is called the forward transform of $f(x, y)$. 

Given $T(u, v)$, we can recover $f(x, y)$ using the inverse transform

$f(x,y) = \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} T(u, v) s(x, y, u, v)$ **(Eq 2-19)**

where $s(x, y, u, v)$ is called the inverse transformation kernel. 

## 3. Intensity Transformations and Spatial Filtering

### 3.1 Basics

Spatial domain processes are based on the expression

$g(x, y) = T[f(x, y)]$ **(Eq 3-1)**

where $T$ is an operator on $f$ defined over a neighborhood of point $(x, y)$

**NOTE**: Approaches whose results depend only on the intensity at a point sometimes are called *point processing techniques*.

### 3.2 Basic Intensity Transformation Functions

**3.2.1 Image Negatives**

$s = L - 1 - r$ **(Eq 3-2)**

**3.2.2 Log Transformations**

$s = c \log(1+r)$ **(Eq 3-3)**

This transform maps a narrow range of low intensity values in the input to a wider range of output levels. Such transformations are used to expand the values of dark pixels in an image, while compressing the higher-level values. The opposite is true for inverse log (exponential)transformation

**3.2.3 Power Law (Gamma) Transformations**

$s = cr^{\gamma}$ **(Eq 3-4)**

Power law curves with fractional values of $\gamma$ map a narrow range of dark input values into a wider range of output values with the opposite being true for higher level of input levels. 

**3.2.4 Piecewise Linear Transformation Functions**

Advantage of PLTF functions is that the form of piecewise functions can be arbitrarilty complex. The main disadvantage of these functions is that their specification requires considerable user input. Two transformations in PLTF are:
1. Contrast Stretching
2. Intensity-level Slicing
3. Bit plane slicing

### 3.3 Histogram Processing

Let $r_{k}$ for $k = 0, 1, 2,...,L - 1$ denote the intensities of an L-level digital image, $f(x, y)$. The unnormalized histogram of $f$ is defined as 

$h(r_{k}) = n_{k}$ **(Eq 3-5)**

where $n_{k}$ is the number of pixels in $f$ with intensity $r_{k}$, and the subdivisions of the intensity scale are called *histogram bins*. 

Normalized histogram of $f$ is defined as:

$p(r_{k}) = h(r_{k}) / MN = n_{k} / MN$ **(Eq 3-6)**

The components of $p(r_[{k})$ are estimates of the probabilities of intensity levels occuring in an image.

### 3.4 Histogram Equalization

Let variable $r$ denote the intensities of an image to be processed. $r$ is in the range $[0, L-1]$. Transformation is given by 
$s = T(r)$. We assume that-

(a) $T(r)$ is a monotonic increasing function in the interval $0 \le r \le L-1$; and
(b) $0 \le T(r) \le L-1$ for $0 \le r \le L-1$

NOTE: A function $T(r)$ is a monotonic increasing function if $T(r_{2}) \ge T(r{1}) for r_{2} > r_{1}$. $T(r)$ is a strictly monotonic inreasing function if $T(r_{2}) > T(r_{1}) for r_{2} > r_{1}$.

For some formulations we use the inverse transformation: 

$r = T^{-1}(s)$ **(Eq 3-7)**
 
in which case we change the condition (a) to:
(a'): $T(r)$ is a strictly monotonic increasing function

The condition in (a) that $T(r)$ be monotonically increasing guarantees that output intensity values will never be less that corresponding input values, thus preventing artifacts created by reversal of intensity. Condition (b) guarantees that the range of output intensities is the same as the input. Condition (a') guarantees that the mappings from s back to r will be one-to-one, thus preventing ambiguities.

$T(r)$ being strictly monotonic guarantees that the inverse mappings will be single values. 

The intensity of an image may be viewed as a random variable in the interval $[0, L-1]$. Let $p_{r}(r)$ and $p_{s}(s)$ denote the PDFs of the intensity values r and s in two different images. A fundamental result from probability theory is that if $p_{r}(r)$ and $T(r)$ are known, then the PDF of the transformed variable s can be obtained as:

$p_{s}(s) = p_{r}(r)|dr / ds|$ **(Eq 3-8)**

A transformation function of particular importance in image processing is:

$s = T(r) = (L-1)\int_0^r p_{r}(w)dw$ **(Eq 3-9)**

where $w$ is a dummy variable of integration. The integration on the right side is the CDF of random variable r. 

By Leibniz's rule in calculus, derivative of a definite integral with respect to its upper limit is the integrand evaluated at the limit. That is:

$\frac{ds}{dr} = \frac{dT(r)}{dr} = (L-1)\frac{d}{dr}[\int_0^r p_{r}(w)dw] = (L-1)p_{r}(w)dw$ **(Eq 3-10)**

Substituting this result for ds/dr in (Eq 3-8), we get

$p_s{s} = \frac{1}{L-1}$ **(Eq 3-11)

We recognize this form as a uniform probability density function. 

For discrete values, we work with probabilities and summations instead of probabiliy density functions and integrations. The discrete form of the transformation in (Eq 3-9) is

$s_{k} = T(r_{k}) = (L-1)\sum_{j=0}^{k} = p_{r}(r_{j})$ **(Eq 3-12)**

Thus, a processed (output image) is obtained by using (Eq 3-12) to map each pixel in the input image with intensity $r_{k}$ into a corresponding pixel level $s_{k}$ in the output image. This is called *histogram equalization*

### 3.5 Histogram Matching or Histogram Specification

There are applications in which histogram equalization is not suitable. It is useful sometimes to be able to specify the shape of the histogram that we wish the processed image to have. This is done by *Histogram Matching*.

Consider continuous intensities $r$ and $z$ with PDFs $p_{r}(r)$ and $p_{z}(z)$, respectively. Here $r$ and $z$ denote the intensity levels of the input and output images. We estimate $p_{r}(r)$ from the given input image and $p_{z}(z)$ is the specified PDF that we wish the output image to have

Let $s$ be a random variable with the property

$s = T(r) = (L-1)\int_0^r p_{r}(w)dw$ **(Eq 3-13)**

Define a function $G$ on variable $z$ with the property

$G(z) = (L-1)\int_0^z p_{z}(v)dv = s$ **(Eq 3-14)**

where $v$ is a dummy variable of integration. It follows from the preceding two equations that $G(z) = s = T(r)$ and therefore, that $z$ must satisfy the conditio

$z = G^{-1}(s) = G^{-1}[T(r)]$ **(Eq 3-15)**

The discrete formulation of (Eq 3-13) is given by:

$s_{k} = T(r_{k}) = (L-1)\sum_{j=0}^{k} = p_{r}(r_{j})$  **(Eq 3-16)**

Then, we compute the transformation function

$G(z_{q}) = (L-1)\sum_{i=0}^{q}p_{z}(z_{i})$ **(Eq 3-17)**

for a value q so that 

$G(z_{q}) = s_{k}$ **(Eq 3-18)**

### 3.7 Fundamentals of Spatial Filtering

Spatial Filtering modifies an image by replacing the value of each pixel by a function of values and its neighbors. If the operation performed on the image pixels is linear, then the filter is called *linear filter.* Otherwise, it's called *nonlinear filter*

**3.7.1 Mechanics of Linear Spatial Filtering**
A linear spatial filter performs a sum-of-products operation between an image $f$ and a filter kernel $w$ also called as window, mask, and template. 

For a kernel of size $ m \times n$, we assume $m = 2a + 1$ and $n = 2b + 1$ where $a$ and $b$ are both nonnegative integers. 

Linear spatial filtering of an image of size M x N with a kernel size of m x n is given by

$g(x, y) = \sum_{s=-a}^{a}\sum_{t=-b}^{b} w(s,t)f(x+s, y+t)$ **(Eq 3-19)**

### 3.8 Spatial Correlation and Convolution

(Eq 3-19) is the mathematical description of correlation. Correlation consists of moving the center of a kernel over an image, and computing the sum of products at each location. The mechanics spatial convolution are the same, except that the correlation kernel is rotated by 180 degrees. 

Few important points about correlation:
- Correlation if a functionof displacement of the filter kernel reltaive to the image. 
- Correlating a kernel $w$ with a function that contains all 0's and a single 1 (impulse signal) yields a copy of $w$, but rotated by 180 degrees. 

In the case of convolution, we have prerotated the kernel by 180 degrees. The result of doing so is that now we have an exact copy of the kernel at the location of the unit impulse. 

Correlation of a kernel $w$ of size m x n with an image $f(x, y)$ is given by

$(w \ast f)(x, y) = \sum_{s=-a}^{a}\sum_{t=-b}^{b} w(s,t)f(x+s, y+t)$ **(Eq 3-20)**

Similarly, convolution of a kernel $w$ of size m x n with an image $f(x, y)$ is denoted by
$(w \ast f)(x, y) = \sum_{s=-a}^{a}\sum_{t=-b}^{b} w(s,t)f(x-s, y-t)$ **(Eq 3-21)**

where the minus sign aligns align the coordinates of $f$ and $w$ when one of the function is rotated by 180 degrees.

**Padding**: If the kernel is of size m x n, we pad the image with a minimum of (m-1)/2 rows of 0's at the top and bottom and (n-1)/2 columns of 0's on the left and right.

If the kernel and image are of sizes m x n and M x N respectively, the padding would have to increase to (m-1) padding elements above and below the image,  and (n-1) elements to the left and right. Under these conditions, the size of the resulting full convolution array will be of size $S_{v}$ and $S_{h}$ where

$S_{v} = m + M - 1$ **(Eq 3-22)**

$S_{h} = n + N - 1$ **(Eq 3-23)**

Sometimes an image is filtered (convolved) sequentially, in stages, using a different kernel in each stage. For example, suppose an image $f$ is filtered with a kernel $w_{1}$, the result filtered with kernel $w_{2}$, that result filtered with a third kernel and son on, for Q stages. Because of commutative property of convolution, this multistage filtering can be done in a single filtering operation $w \ast f$, where

$w = w1 \ast w2 \ast ... \ast w_{Q}$

The size of $w$ is obtained from the sizes of the individual kernels by successive applications of (Eq 3-22) and (Eq 3-23). If all the kernels are of size m x n, it follows from these equations that $w$ will be of size:

$W_{v} = Q * (m-1) + m$ **(Eq 3-24)**

$W_{h} = Q * (n-1) + n$ **(Eq 3-25)**

### 3.9 Smoothing (Lowpass) Spatial Filters

Smoothing filters are used to reduce sharp transitions in intensity. Noise reduction is one of the common applications of smoothing since noise typically consists of sharp transitions in intensity. 

Convolving a smoothing kernel with an image blurs the image, with the degree of blurring being determined by the size of the kernel and the values of its coefficients.

**3.9.1 Box Filter Kernels**
Simplest, separable, lowpass filter kernel whose coefficients have the same value (1). A box kernel is a m x n array of 1's, with a normalizing constant, whose value is 1 divided by the sum of the values of the coefficients. This normalization has two purposes. One, the average value of an area of constant intensity would equal that intensity in the filtered image. Second, normalizing the kernel in this way prevents introducing a bias during filtering; that is the sum of the pixels in the original image and the filtered image will be the same. 

**3.9.2 Lowpass Gaussian Filter Kernels**
Kernels in applications where high level detail needs to preserved are required to be circulary symmetric, that is the blurring should not happen in perpendicular directions. Gaussian kernels of the form

$w(s, t) = G(s, t) = Ke^{- (s^2 + t^2) / (2\sigma^{2})}$ **(Eq 3-26)

are the only circulary symmetric kernels that are also separable.

By letting  $r = [s^2 + t^2]^{1/2}$, we can write

$G(r) = Ke^{- r^2 / (2\sigma^{2})}$ **(Eq 3-27)**

where $r$ is the distance from the center to any point on function $G$.

**3.9.3 Order Statistic (Nonlinear) Filters**

These are filters whose response is based on ordering (ranking) of the pixels contained in the area encompassed by the kernel.Smoothing is achieved by replacing the value of the center pixel with the value determined by the ranking result. Mean, median are examples. 

### 3.10 Sharpening (Highpass) Spatial Filters

Sharpening highlights transitions in intensity. We saw that image blurring could be accomplished in the spatial domain by pixel averaging. Because averaging is analogous to integration, it is logical to conclude that sharpening can be accomplished by spatial differentiation. The strength of the reponse of a derivative operator is proportional to the magnitude of the intensity discontinuity at the point at which the operator is applied. 

Derivatives of a digital function are defined in terms of differences. Basic definition of first order derivative is:

$\partial{f} / \partial{x} = f(x+1) - f(x)$ **(Eq 3-28)**

$\partial^{2}{f} / \partial{x}^2 = f(x-1) + f(x+1) - 2f(x)$ **(Eq 3-29)**

**3.10.1 Using the Second Derivative for Image Sharpening - The Laplacian**

The simplest isotropic derivative operator is the Laplacian. For a function of two variables, it is defined as

$\nabla^2{f} = \partial^{2}{f} / \partial{x}^2 + \partial^{2}{f} / \partial{y}^2$ **(Eq 3-30)**

where $partial^{2}{f} / \partial{x}^2 = f(x+1, y) + f(x-1, y) - 2f(x,y)$ **(Eq 3-31)** and

$partial^{2}{f} / \partial{y}^2 = f(x, y+1) + f(x, y-1) - 2f(x,y)$ **(Eq 3-32)**

Thus

$\nabla^2{f(x, y)} = f(x+1, y) + f(x-1, y) + f(x, y+1) + f(x, y-1) - 4f(x, y)$ **(Eq 3-33)**

The basic way in which we use the Laplacian for image sharpening is 

$g(x, y) = f(x, y) + c[\nabla^2{f(x, y)}]$ **(Eq 3-34)**

**Unsharp Masking and Highboost Filtering**

Process of unsharp masking:
- Blur the original image
- Subtract the blurred image from the original (resulting difference is called the mask)
- Add weighted portion of the mask (k) to the original

When:
- k = 1, we have unsharp masking
- k > 1, we have highboost filtering

**3.10.1 Using the First Order Derivative for Image Sharpening - The Gradient**

The gradient of an image $f$ at coordinates (x, y) is defined as a 2D column vector

$\nabla{f} = grad(f) = \begin{bmatrix} g_x \\ g_y \end{bmatrix}$ **(Eq 3-35)**

where $g_x = \partial{f}/\partial{x}$ and $g_y = \partial{f}/\partial{y}$

This vector has the important geometrical property that it points in the direction of the greatest rate of change of at location $(x, y)$

The magnitude of $\nabla{f}$

$M(x, y)$ = ||$\nabla{f}$|| = mag($\nabla{f}$) = $[g_x^{2} + g_y^{2}]^{1/2}$ **(Eq 3-36)**

is the value at $(x, y)$ of the rate of change in the direction of the gradient vector.

### 3.11 Highpass, Bandreject, and Bandpass Filters from Lowpass Filters

Bandreject filter can be obtained by adding lowpass and highpass filters. Bandpass filter can be obtained by subtracting bandreject filter from a unit impulse.

## 4. Filtering in the Frequency Domain

Fourier transform of the convolution of two functions in the spatial domain is equal to the product in the frequency domain of the Fourier transforms of the two functions.

### 4.1 2D Discrete Fourier Transform and it's Inverse


$F(u, v) = \sum_{x=0}^{M-1}\sum_{y=0}^{N-1} f(x, y)e^{-2j\pi(ux/M + vy/N)}$ **(Eq 4-1)**

where $f(x, y)$ is a digital image of size M x N. The above equation must be evaluated for values of the discrete variables $u$ and $v$ in the ranges $u$ = 0, 1, 2, ..., M-1 and $v$ = 0, 1, 2, ..., N-1

Given the transform $F(u, v)$, we can obtain $f(x, y)$ by using inverse DFT:

$f(x, y) = (1/MN)sum_{u=0}^{M-1}\sum_{v=0}^{N-1} F(u,v)e^{2j\pi(ux/M + vy/N)}$ **(Eq 4-2)**

for $x$ = 0, 1, 2, ... , M-1 and $y$ = 0, 1, 2, ... , N-1

### 4.2 The 2D Discrete Convolution Theorem

$(f \ast h)(x, y)  = \sum_{m=0}^{M-1}\sum_{n=0}^{N-1} f(m, n)h(x-m, y-n)$ **(Eq 4-3)**

The 2D convolution theorem is given by:

$(f \ast h)(x, y) <-> (F \cdot H)(u, v)$**(Eq 4-4)**

and conversely $(f \cdot h)(x, y) <-> (1/MN)(F \ast H)(u, v)$ **(Eq 4-5)**

where F and H are the Fourier transforms of f and h, obtained using (Eq 4-1)

(Eq 4-4) states that the Fourier transform of the spatial convolution of f and h, is the product of their transforms. (Eq 4-4) is the foundation of linear filtering in the frequency domain and is the basis for all the filtering techniques discussed in this chapter. It is the tie that establishes the equivalence between the spatial and frequency domain filtering.

We are interested in the results of convolution in the spatial domain, where we analyze images. However, the convolution theorem tells us that we have two ways of computing the spatial convolution of two function. We can do it directly by performing convolution in the spatial domain or, we can compute the Fourier transforms of the two functions and multiply the transforms, and compute the inverse Fourier transform.

Because we are dealing with discrete quantities, computation of Fourier transform is carried out using DFT algorithm. *This automatically implies periodicity*. which means that when we take the inverse Fourier transform of the product of the transforms, we would get a circular (i.e. periodic) convolution. The question then arises: *Under what conditions will the direct spatial approach and the inverse Fourier transform method yield the same approach?* Going with the approach of computing Fourier transforms of the functions f and h, multiplying them and then using inverse Fourier transform results into a convolution that is also periodic and gives a *wraparound error*. This error results when the closeness of the periods is such that they interfere with each other. 

The solution to this is simple. Consider two functions, $f(x)$ and $h(x)$ composed of A and B samples, respectively. It can be shown that if we append zeros to both functions so that they have the same length P, then wraparound is avoided by choosing

$P \ge A + B - 1$ **(Eq 4-6)**

This process is called zero padding. The same can be applied in 2D. Consider $f(x, y)$ and $h(x, y)$ be two image arrays of sizes A x B and C x D pixels, respectively. Wraparound error in their convolution can be avoided by padding these functions with zeros as follows:

$ f_p (x, y) = \begin{Bmatrix} f(x, y) & 0 \le x \le A-1 and 0 \le y \le A-1 \\ 0 & A \le x \le P or B \le y \le Q\end{Bmatrix}$ **(Eq 4-7)**

and

$ h_p (x, y) = \begin{Bmatrix} h(x, y) & 0 \le x \le c-1 and 0 \le y \le D-1 \\ 0 & C \le x \le P or D \le y \le Q\end{Bmatrix}$ **(Eq 4-8)**

with $P \ge A + C - 1$ **(Eq 4-9)**

and 

$Q \ge B + D - 1$ **(Eq 4-10)**

The resulting images are of size $P \times Q$. 

As a rule, DFT algorithms tend to execute faster with arrays of even size, so it is good practice to selet P and Q as the smallest evven integers that satisfy preceding equations. This means that P and Q are selected as:

$ P = 2M$ **(Eq 4-11)**

and 

$Q = 2N$ **(Eq 4-12)**

### 4.3 Frequency Domain Filtering Fundamentals

Given a padded image, $f(x, y)$, of size $ P \times Q$ pixels, the basic filtering equation in which we are interested has the form

$g(x, y) = Real(\mathcal{F}^{-1}[H(u,v)F(u,v)])$ **(Eq 4-13)**

where $\mathcal{F}^{-1}$ is the IDFT, $F(u, v)$ is the DFT of the input image, $H(u, v)$ is a filter transfer function (filter) and $g(x, y)$ is the output image. The product of $H(u, v)F(u,v)$ is formed using elementwise multiplication.The task of specifying $H(u, v)$ is simplified considerably by using functions that are symmetric about their center, which requires $F(u, v)$ to be centered also. This is accomplished by multiplying the input image $f(x, y)$ by $(-1)^{x+y}$ prior to computing it's transform.

Thus far, the discussion has centered on padding the input image. However, (Eq 4-13) also involves a filter transfer function that can be specified either in the spatial or in the frequency domain. But padding is done in the spatial domain, which raises an important question about relationship between spatial padding and filter functions specified directly in the frequency domain.

It would be reasonable to conclude that the way of handling padding of a frequency domain transfer function construct the function the same size as the unpadded image, compute the IDFT of the function to obtain the corresponding spatial representation, pad the representation in the spatial domain, and then compute its DFT to return to the frequency domain. 

**4.3.1 Steps for Filtering in the Frequency Domain**

1. Given an input image $f(x, y)$ of size $M \times N$, obtain the padding sizes P and Q (4-11) and (4-12)
2. Form a padded image $f_{p}(x,y)$ of size $P \times Q$ using zero, mirror, or replicate padding (In *mirror* padding, values outside the boundary of the image are obtained by mirror-reflecting the image across its border; and *replicate* padding, in which values outside the boundary are set equal to the nearest image border value)
3. Multiply $f_{p}(x,y)$ by $(-1)^{x+y}$ to center the Fourier transform on the $P \times Q$ frequency triangle.
4. Compute the DFT, $F(u,v)$, of the image from step 3
5. Construct a real symmetric filter transfer function, $H(u,v)$ of size $P \times Q$ with center $(P/2, Q/2)$
6. Form the product $G(u,v) = H(u,v)F(u,v)$ using elementwise multiplication
7. Obtain the filtered image by computing IDFT of $G(u,v)$: $g_{p}(x,y) = (real[\mathcal{F}^{-1}(G(u,v))](-1)^{x+y}$
8. Obtain the final filtered result, $g(x,y)$ of the same size as the input image, by extracting $M \times N$ region from the top left of the quadrant $g_{p}(x,y)$

### 4.4 Image Smoothing using Lowpass Frequency Domain Filters

**4.4.1 Ideal Lowpass Filter**

$H(u,v) = \begin{Bmatrix}1 & if D(u,v) \le D_0 \\
                0  & if D(u,v) > 0\end{Bmatrix}$ **(Eq 4-14)**
                
$D_0$ is the cutoff frequency and $D(u,v)$ is the distance between a point $(u, v)$ in the frequency domain and the center of $P \times Q$ frequency triangle.

**4.4.2 Gaussian Lowpass Filter**

$H(u,v) = e^{-D^{2}(u,v)/2\sigma^{2}}$ **(Eq 4-15)**

where $D(u,v)$ is the distance from the center of the $P \times Q$ frequency rectangle to any point $(u,v)$ contained in the rectangle.

**4.4.3 Butterworth Lowpass Filter**

$H(u, v) = 1 / 1+[D(u,v) / D_0]^{2n}$ **(Eq 4-16)**

where $n$ is the order of the filter.

### 4.5 Image Smoothing using Highpass Frequency Domain Filters

Subtracting a lowpass filter transfer function from 1 yields the corresponding highpass filter transfer function in the frequency domain

$H_{HP} = 1 - H_{LP}(u,v)$ **(Eq 4-17)**

- Ideal 
$H(u,v) = \begin{Bmatrix}0 & if D(u,v) \le D_0 \\
                1  & if D(u,v) > 0\end{Bmatrix}$ **(Eq 4-18)**
                
- Gaussian
$H(u,v) = 1 - e^{-D^{2}(u,v)/2\sigma^{2}}$ **(Eq 4-19)**

- Butterworth
$H(u, v) = 1 / 1+[D_0 / D(u,v)]^{2n}$ **(Eq 4-20)**

### 4.6 The Laplacian in the Frequency Domain

$H(u,v) = -4\pi^{2}(u^{2} + v^{2})$ **(Eq 4-21)** or

with respect to the center of frequency triangle,

$H(u,v) = -4\pi^{2}D^{2}(u,v)$ **(Eq 4-22)**

Using this transfer function, the Laplacian of the image is obtained in the familiar manner:

$\nabla^{2}f(x, y) = \mathcal{F}^{-1}[H(u,v)F(u,v)]$ **(Eq 4-23)**

Here c = -1 since $H(u, v)$ is negative. Computing (4-23) introduces DFT scaling factors that can be several orders of magnitude larger than the maximum comparable ranges. The easiest way to handle this problem is to normalize the values of $f(x, y)$ to the range [0,1] (before computing its DFT) and divide $\nabla^{2}f(x, y)$ by its maximum value, which will bring it to the approximate range [-1, 1]. 

### 4.7 Unsharp Masking, Highboost Filtering, and High Frequency-Emphasis Filtering

The mask is defined by

$g_{mask} = f(x,y) - f_{LP}(x,y)$ **(Eq 4-24)**

where $f_{LP}(x,y) = \mathcal{F}^{-1}[H(u,v)F(u,v)]$ **(Eq 4-25)**

Then,

$g(x,y) = f(x,y) + kg_{mask}(x, y)$ **(Eq 4-26)**

This expression defines unsharp masking when k = 1 and high-boost filtering when k > 1.

We can express this result in terms of highpass filter as

$g(x,y) = \mathcal{F}^{-1}[[1 + kH_{HP}(u,v)]F(u,v)]$ **(Eq 4-27)**

This expression is called *high-frequency emphasis filter transfer function*. $k$ gives control over the proportion of high frequencies that influences the final result. 

### 4.8 Homomorphic Filtering

$g(x,y) = i_0(x,y)r_0(x,y)$ **(Eq 4-27)**

where

$i_0 = e^{i'(x,y)}$
$r_0 = e^{r'(x,y)}$

where

$i'(x,y) = \mathcal{F}^{-1}[H(u,v)F_i(u,v)]$
$r'(x,y) = \mathcal{F}^{-1}[H(u,v)F_r(u,v)]$

where 

$F_i(u,v) = ln i(x,y)$
$F_r(u,v) = ln r(x,y)$

### 4.9 Bandreject and Bandpass Filters

$W$: width of the band
$C_0$: center of the band

- Ideal Bandreject Filters

$H(u,v) = \begin{Bmatrix}0 & if C_0 - (W/2) \le D_0 \le C_0 + (W/2) \\
                1  & otherwise\end{Bmatrix}$ **(Eq 4-28)**

- Gaussian Bandreject filter

$H(u,v) = 1 - e^({(D^2(u,v) - C_0^{2}) / D(u,v)W)})^2 $ **(Eq 4-29)**

- Butterworth Bandreject Filter

$H(u,v) = 1 / 1 + ({D(u,w)W / D^{2}(u,v) - C_0^{2}})^{2n}$ **(Eq 4-30)**

For bandpass filters

$H_{BP}(u,v) = 1 - H_{BR}(u,v)$ **(Eq 4-31)**

### 4.10 Notch Filters

A notch filter rejects (or passes) frequencies in a predefined neighborhood of the frequency rectangle. The general form is 

$H_{NR}(u,v) = \prod_{k=1}^{Q} H_k(u, v)H_{-k}(u,v)$ **(Eq 4-32)**

## 5. Image Restoration and Reconstruction

### 5.1 Introduction

We model image degradation as an operator $\mathcal{H}$ that, together with an additive noise term, operates on an input image $f(x, y)$. 

Given $g(x,y)$, some knowledge about $\mathcal{H}$, and some knowledge about the additive noise term $\eta(x, y)$, the objective of restoration is to obtain an estimate $\hat{f}(x,y)$ of the original image. 

We want the estimate to be as close as possible as possible to the original image and, in general, the more we want to know about $\mathcal{H}$ and $\eta$, the closer $\hat{f}(x,y)$ will be to $f(x, y)$

If $\mathcal{H}$ is a linear, position-invariant operator, then the degraded image is given in the spatial domain by

$g(x, y) = (h \ast f)(x,y) + \eta(x, y)$ **(Eq 5-1)**

where $h(x, y)$ is the spatial degradation of the degradation function. Equivalent of (5-1) in the frequency domain is

$G(u, v) = H(u, v)F(u, v) + N(u, v)$ **(Eq 5-2)**

### 5.2 Noise Models

We create a noise image by generating an array whose intensity values are rendom numbers with a specified probability density function.

1. Gaussian Noise
2. Rayleigh Noise
3. Erland (Gamma Noise)
4. Exponential Noise
5. Uniform Noise
6. Salt and Pepper Noise

### 5.3 Restoration in the presence of noise only - Spatial Filtering

When an image is degraded only by additive noise (5-1) becomes

$g(x, y) = f(x,y) + \eta(x, y)$ **(Eq 5-3)**

and

$G(u, v) = F(u, v) + N(u, v)$ **(Eq 5-4)**

The noise terms generally are unknown, so subtracting them from $g(x, y)$ to obtain $f(x, y)$ typically is not an option. In case of periodic function, sometimes it is possible to estimate $N(u, v)$ from the spectrum of $G(u, v)$. In this case $N(u, v)$ can be subtracted from $G(u, v)$ to obtain an estimate of the original image. But this is an exception rather than a rule.

Spatial filtering is the method of estimating $f(x, y)$ (denoising) in situations when only additive random noise is present.

Periodic noise can be analyzed and filtered quite effectively using frequency domain techniques. The basic idea is that periodic noise appears as concentrated bursts of energy in the Fourier transform, at locations corresponding to the frequencies of the periodic interference. The approach is to use a selective filter (bandreject, bandpass, notch) to isolate the noise. The tool of choice is the notch filter.

## 6. Color Image Processing

### 6.1 Color Fundamentals

Characterization of light is central to the science of color. Chromatic light spans the electromagnetic spectrum from approximately 400 to 700 nm. Three basic quantities used to describe the quality of chromatic light are:
- Radiance: total amount of energy that flows from the light source, usually measured in Watts(W)
- Luminance: measure of the amount of energy that an observer perceives from a light source, measured in lumens(L)
- Brightness

The characteristics generally used to distinguish one color from another are:
- Brightness
- Hue: attribute associated with the dominant wavelength in a mixture of light waves. Hue represents dominant color as perceived by an observer
- Saturation: relative purity of the amount of white light mixed with a hue

Hue and saturation taken together are called chromaticity and therefore, a color may be characterized by its brightness and chromaticity. The amount of reg, green, and blue needed to form any particular color are called the tristimulus values, and are denoted by $x$, $Y$, and $Z$ respectively.

### 6.2 Color Models

The purpose of color models is to facilitate the specification of colors in some standard way. A color model is a specification of:
- a coordinate system
- a subspace within that system, such that each color in the model is represented by a single point contained in that subspace

Known color models are:
- RGB
- CMY and CMYK
- HSI

### 6.3 Pseudocolor Image Processing

Pseudocolor image processing consists of assigning colors to gray values based on specific criterion. 

**6.3.1 Intensity Slicing and Color Coding**

If an image is interpreted as a 3D function, instensity slicing and color coding can be viewed as one of placing planes parallel  to the coordinate plane of the image; each plane then "slices" the function in the area of intersection.



The above figure shows an example of using a plane at $f(x, y) = l_i$ to slice the image intensity function into two levels. If a different color is assigned to each side of the plane in the above figure, any pixel whose intensity level is above the plane will be coded with one color, and any pixel below the plane will be coded with another. Levels that lie on the plane itself may be arbitrarily assigned one of the two colors, or they could be given a third color to highlight all the pixels at that level. This can be generalized as follows:

Let $[0, L-1]$ represent the grayscale, let level $l_0$ represent black ($f(x,y) = 0$) and level $l_{L-1}$ represent white ($f(x,y) = L-1$). Suppose that $P$ planes perpendicular to the intensity axis are defined at levels $l_1, l_2, ..., l_P$. Then assuming that $0 < P < L-1$, the P planes partition the grayscale into $P+1$ intervals, $I_1, I_2, ..., I_P+1$. Intensity to color assignments at each pixel location $(x, y)$ are made according to the equation

if $f(x, y) \in I_k$, let $f(x, y) = c_k$ **(Eq 6-1)**

where $c_k$ is the color associated with the kth intensity interval $I_k$, defined by the planes at $l = k-1$ and $l=k$

**6.3.2 Intensity to Color Transformations**

The idea underlying this approach is to perform three independent transformations on the intensity of input pixels. The three results are then fed separately into the red, green, and blue channels of a color monitor. This method produces a composite image whose color content is modulated by the nature of the transformation functions. 

### 6.4 Full Color Image Processing

Full color image processing approaches fall into two major categories. In the first category, we process each grayscale component image individually, then form a composite color image from the individually processed components. In the second category, we work with color pixels directly. 

Because full color images have at least three components, color pixels are vectors. For example, in the RGB system, each color print can be interpreted as a vector extending from the origin to that point in the RGB coordinate system. 

Let $c$ represent an arbitrary vector in RGB color space:

$c = \begin{bmatrix} c_R \\ c_G \\ c_B \end{bmatrix} = \begin{bmatrix} R \\ G \\ B \end{bmatrix}$ **(Eq 6-2)**

### 6.5 Color Transformations

The general expression for color transformation is:

$s_i = T_i(r_i) i = 1, 2, ..., n$ **(Eq 6-3)**

where n is the total number of component images, $r_i$ are the intensity values of the input component images, $s_i$ are the spatially corresponding intensities in the outpu component images, and $T_i$ are a set of transformation or color mapping functions that operate on $r_i$ to produce $s_i$. 

### 6.6 Color Slicing

Highlighting a specific range of colors in an image is useful for separating objects from their surroundings. The basic idea is to: (1) display the colors of interest or (2) use the region defined by the colors as a mask for further processing. 

One of the simplest ways to "slice" a color image is to map the colors outside some range of interest into a nonprominent neutral color.

### 6.7 Histogram Processing of Color Images

It is generally unwise to histogram equalize the component images of a color image independently. This results in erroneous color. A more logical approach is to spread the color intensities uniformly, leaving the colors themselves unchanges. 


### 6.8 Color Image Smoothing

Let $S_{xy}$ denote the set of coordinates defining a neighborhood centered at $(x, y)$ in an RGB color image. The average of the RGB component vectors in this neighborhood is:

$c(x, y) = (1/K) \sum_{(s, t) \in S_{xy}} c(s, t)$ **(Eq 6-4)**

where $c(s, t)$ follows from (6-2)

### 6.9 Color Image Sharpening

Color Image Sharpening is done using the Laplacian

$ \nabla^{2}[c(x, y)] = \begin{bmatrix} \nabla^{2}R(x, y) \\ \nabla^{2}G(x, y) \\ \nabla^{2}B(x, y) \end{bmatrix}$ **(Eq 6-5)**

## 7. Wavelets and Other Image Transforms

### 7.1 Introduction

The discrete Fourier transform is a member of an important class of linear transformations that include the Hartley, sine, cosine, Walsh-Hadamard, Slant, Haar, and Wavelet transforms. These transforms decompose functions into weighted sums of orthogonal or biorthogonal basis functions, and can be studied using the tools of linear algebra and functional analysis.

When approached from this point of view, images are vectors in the vector space of all images. Basis functions determine the nature and usefulness of image transforms. Transforms are the coefficients of linear expansions. And for a given image and transform, both the orthogonality of the basis functions and the coefficients of the resulting transform are computed using the inner products. All of an image's transforms are equivalent in the sense that they contain the same information and total energy. They are reversible and differ only in the way that the information and energy is distributed among the transform's coefficients. 

### 7.2 Fourier related Transforms

1. Hartley Transform
2. Cosine Transform
3. Sine Transform
4. Walsh-Hadamard Transform
5. Slant Transform
6. Haar Transform

### 7.3 Wavelet Transforms

Multiresolution theory incorporates and unifies techiques from a variety of disciplines, including subband coding from signal processing, quadrature mirror filtering from digital speech recognition, and pyramidal image processing. As the name suggests, it is concerned with the representation and analysis of signals at more than one resolution. A scaling function is used to create a series of approximations of a function or image, each differing by a factor of 2 in resolution from its nearest neighboring approximations, and complementary functions, called *wavelets* are used to encode the differences between adjacent approximations. 

## 8. Morphological Image Processing

### 8.1 Erosion

Morphological expressions are written in terms of structuring elements and a set A, of foreground pixels, or in terms of structuring elements and an image I, that contains A. If B is a structuring element, erosion of A by B is the set of all points z such that B, translated by z, is contained in A. 

### 8.2 Dilation

Dilation of A by B is the set of all displacements, z, such that the foreground elements of B overlap at least one element of A. 

### 8.3 Opening

Opening smoothes the contour of an object, breaks narrow isthmuses, and eliminates thin protrusions. The opening of A by B is the union of all the translations of B so that B fits entirely in A. 

### 8.4 Closing

Closing has a similar geometric interpretation as opening, excepts that now we translate B outsite A. The closing is then the complement of the union of all translations of B that do not overlap A. 

### 8.5 Hit or Miss Transform

HMT is a basic tool for shape detection. Let I be a binary image composed of foreground (A) and background pixels, respectively. HMT utilizes two structuring elements: B1 for detecting shapes in the foreground, and B2, for detecting shapes in the background. 

### 8.6 Basic Morphological Algorithms

**8.6.1 Boundary Extraction**: The boundary of a set A of foreground pixels can be obtained by first eroding A by a suitable SE, and then performing set difference between A and its erosion. 

**8.6.2 Extraction of Connected Components**: Let A be a set of foreground pixels consisting of one or more connected components, and form an image X0 (of the same size as I) whose elements are 0's, except at each location known to correspond to a poin in each connected component  in A, which we set to 1's. 

**8.6.3 Convex Hull**: A set S of points in the Euclidean plane is said to be convex if and only if a straight line segment joining any two points in S lies directly within S. The convex hull H, of S is teh smallest convex set containing S.
