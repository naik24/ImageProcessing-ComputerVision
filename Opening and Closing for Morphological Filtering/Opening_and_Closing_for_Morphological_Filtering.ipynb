## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Opening and Closing for Morphological Filtering

```python
# loading the image
noisy_fingerprint = cv2.imread('noisy_fingerprint.tif')

# structuring element
structuring_element = np.ones((3, 3))

# opening of the original image (dilation of erosion)
noisy_fingerprint_opening = cv2.morphologyEx(noisy_fingerprint, cv2.MORPH_OPEN, structuring_element)

# closing of the opened image
noisefree_fingerprint = cv2.morphologyEx(noisy_fingerprint_opening, cv2.MORPH_CLOSE, structuring_element)

# displaying the results
plt.figure(figsize = (50, 50))
plt.subplot(1, 3, 1); plt.axis('off'); plt.imshow(noisy_fingerprint); plt.title('Noisy Fingerprint')
plt.subplot(1, 3, 2); plt.axis('off'); plt.imshow(noisy_fingerprint_opening); plt.title('Opening Operation')
plt.subplot(1, 3, 3); plt.axis('off'); plt.imshow(noisefree_fingerprint); plt.title('Noisefree Fingerpring (Closing of Opened Image)')
```
