## Importing Libraries

```python
import cv2
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Original Image

```python
pattern = cv2.imread('pattern.tif')
pattern = cv2.cvtColor(pattern, cv2.COLOR_BGR2GRAY)
h, w = pattern.shape[:2]
cv2_imshow(pattern)
```

## Gaussian Noise

```python
# defining gaussian noise parameters
mu = 0
var = 0.1
sigma = 400 * var

# defining gaussian noise
gaussian_noise = np.random.normal(mu, sigma, size = (h, w))

# adding gaussian noise to image
pattern_gaussian = pattern + gaussian_noise
cv2_imshow(pattern_gaussian)
```

## Rayleigh Noise

```python
# defining rayleigh noise
rayleigh_noise = np.random.rayleigh(40, size = (h, w))

# adding rayleigh noise to the image
pattern_rayleigh = pattern + rayleigh_noise
cv2_imshow(pattern_rayleigh)
```

## Erlang (Gamma) Noise

```python
# defining erlang noise
erlang_noise = np.random.gamma(2, scale = 20, size = (h, w))

# adding erlang noise to the image
pattern_erlang = pattern + erlang_noise
cv2_imshow(pattern_erlang)
```

## Uniform Noise

```python
# defining uniform noise
uniform_noise = np.random.uniform(low = 0, high = 100, size = (h, w))

# adding uniform noise to the image
pattern_uniform = pattern + uniform_noise
cv2_imshow(pattern_uniform)
```

## Exponential Noise

```python
# defining exponential noise
exponential_noise = np.random.exponential(scale = 40, size = (h, w))

# adding exponential noise to image
pattern_exponential = pattern + exponential_noise
cv2_imshow(pattern_exponential)
```

```python
# defining salt and pepper noise
probability = 0.2
num_pixels = int(h * w * probability)

# allocating memory for salt and pepper noise image
pattern_saltpepper = np.copy(pattern)

# creating salt noise
salt_noise_coords = [np.random.randint(0, h) for _ in range(num_pixels)], \
                 [np.random.randint(0, w) for _ in range(num_pixels)]
pattern_saltpepper[salt_noise_coords] = 229

# creating pepper noise
pepper_coords = [np.random.randint(0, h) for _ in range(num_pixels)], \
                   [np.random.randint(0, w) for _ in range(num_pixels)]
pattern_saltpepper[pepper_coords] = 0

cv2_imshow(pattern_saltpepper)
```
