## Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the image and generating histogram

```python
# loading the image
pollen = cv2.imread('pollen.tif')
pollen = cv2.cvtColor(pollen, cv2.COLOR_BGR2GRAY)

plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(pollen, cmap = 'gray')

# generating histogram
plt.subplot(1, 2, 2)
plt.title("Histogram")
plt.hist(pollen.ravel(), bins = 256, range = [0,256])
plt.show()
```

## Histogram Equalization

```python
pollen_histEq = cv2.equalizeHist(pollen)

plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(pollen_histEq, cmap = 'gray')

# generating histogram
plt.subplot(1, 2, 2)
plt.title("Equalized Histogram")
plt.hist(pollen_histEq.ravel(), bins = 256, range = [0,256])
plt.show()
```
