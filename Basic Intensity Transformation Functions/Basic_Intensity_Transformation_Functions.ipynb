## Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

## Image Negatives

```python
# importing the image
breast_xray = cv2.imread('breast_digital_xray.tif')

# getting the maximum intensity value in the image
max_intensity = np.amax(breast_xray)

# subtracting all the intensities from the max_intensity to get negative image
breast_xray_negative = max_intensity - breast_xray

# plotting the results
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(breast_xray)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title('Negative Image')
plt.imshow(breast_xray_negative)
```

## Log Transformations

```python
# loading the image
fourier_spec = cv2.imread('fourier_spectrum.tif')
fourier_spec = cv2.cvtColor(fourier_spec, cv2.COLOR_BGR2GRAY)
# defining variables for log transformation
c = 1

# performing log transformation on the original image
fourier_spec_log_transformed = np.uint8(c * np.log(1 + fourier_spec))

# plotting the results
# plotting the results
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(fourier_spec, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title('Log Transformed')
plt.imshow(fourier_spec_log_transformed, cmap = 'gray')
```

## Power Law Transformations

```python
# loading the image
fractured_spine = cv2.imread('fractured_spine.tif')
fractured_spine = cv2.cvtColor(fractured_spine, cv2.COLOR_BGR2GRAY)

# defining the parameters for power law transformation
c = 1
gamma = [0.6, 0.4, 0.3]

plt.subplot(2, 2, 1)
plt.axis('off')
plt.title("Original Image")
plt.imshow(fractured_spine, cmap = 'gray')

for g in gamma:
  plt.subplot(2, 2, gamma.index(g) + 2)
  plt.axis('off')
  plt.title(f"gamma  = {g}")
  fractured_spine_power_transformed = c * (fractured_spine ** g)
  plt.imshow(fractured_spine_power_transformed, cmap = 'gray')
```

## Contrast Stretching

```python
# loading the input image
pollen = cv2.imread('pollen.tif')
pollen = cv2.cvtColor(pollen, cv2.COLOR_BGR2GRAY)

# thresholding
threshold_value = (np.amax(pollen) + np.amin(pollen)) / 2
ret, pollen_contrast_stretched = cv2.threshold(pollen,threshold_value,255,cv2.THRESH_BINARY)

# plotting the results
plt.subplot(1, 2, 1)
plt.axis("off")
plt.title("Original Image")
plt.imshow(pollen, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.axis("off")
plt.title("Contrast Stretched")
plt.imshow(pollen_contrast_stretched, cmap = 'gray')
```
