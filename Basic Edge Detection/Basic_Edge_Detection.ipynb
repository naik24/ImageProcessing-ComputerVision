## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Original Image

```python
# loading the original image
building = cv2.imread('building.tif')
building = cv2.cvtColor(building, cv2.COLOR_BGR2GRAY)

# smoothing the original image
kernel = np.ones((5, 5)) / 25
building = cv2.filter2D(building, -1, kernel)
```

## Roberts Filter

```python
# defining the roberts filter
roberts_gx = np.array([[-1, 0], [0, 1]])
roberts_gy = np.array([[0, -1], [1, 0]])

# applying Roberts filter on the original image
roberts_edgeDetection = cv2.filter2D(building, -1, roberts_gx) + cv2.filter2D(building, -1, roberts_gy)

# displaying the results
plt.figure(figsize = (10, 20))
plt.subplot(1, 2, 1); plt.axis('off'); plt.title('Original Image'); plt.imshow(building, 'gray')
plt.subplot(1, 2, 2); plt.axis('off'); plt.title('Roberts Edge Detection'); plt.imshow(roberts_edgeDetection, 'gray')
```

## Prewitt Edge Detection

```python
# defining the Prewitt gradient filters
prewitt_gx = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitt_gy = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

# applying Prewitt filter on the original image
prewitt_edgeDetection = cv2.filter2D(building, -1, prewitt_gx) + cv2.filter2D(building, -1, prewitt_gy)

# displaying the results
plt.figure(figsize = (10, 20))
plt.subplot(1, 2, 1); plt.axis('off'); plt.title('Original Image'); plt.imshow(building, 'gray')
plt.subplot(1, 2, 2); plt.axis('off'); plt.title('Prewitt Edge Detection'); plt.imshow(prewitt_edgeDetection, 'gray')
```

## Sobel Edge Detection

```python
# defining the Sobel gradient kernels
sobel_gx = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
sobel_gy = np.array([[-1, 0, -1], [-2, 0, 2], [-1, 0, 1]])

# applying Sobel filters on the original image
sobel_edgeDetection = cv2.filter2D(building, -1, sobel_gx) + cv2.filter2D(building, -1, sobel_gy)

# displaying the results
plt.figure(figsize = (10, 20))
plt.subplot(1, 2, 1); plt.axis('off'); plt.title('Original Image'); plt.imshow(building, 'gray')
plt.subplot(1, 2, 2); plt.axis('off'); plt.title('Sobel Edge Detection'); plt.imshow(sobel_edgeDetection, 'gray')
```
