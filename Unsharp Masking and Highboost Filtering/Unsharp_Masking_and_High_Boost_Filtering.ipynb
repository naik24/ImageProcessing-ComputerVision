## Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

## Unsharp Masking and Highboost Filtering

```python
# laoding the image
text = cv2.imread('dipxe_text.tif')
#text = cv2.cvtColor(text, cv2.COLOR_BGR2GRAY)

# blurring the original image using Gaussian Blur
text_blurred = cv2.GaussianBlur(text, ksize = (31, 31), sigmaX = 5)

# obtaining the mask (subtracting the blurred image from the original image)
mask = cv2.subtract(text, text_blurred)

# unsharp masking
unsharp_masking = cv2.add(text, mask)

# highboost filtering
highboost_filtering = cv2.add(text, 6 * mask)

# plotting the results
plt.subplot(1, 3, 1)
plt.axis('off')
plt.title("Original Image")
plt.imshow(text)

plt.subplot(1, 3, 2)
plt.axis('off')
plt.title("Unsharp Masking")
plt.imshow(unsharp_masking)

plt.subplot(1, 3, 3)
plt.axis('off')
plt.title("HighBoost Filtering")
plt.imshow(highboost_filtering)
```
