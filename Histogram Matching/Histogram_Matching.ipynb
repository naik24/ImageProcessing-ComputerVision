## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
from skimage import exposure
from google.colab.patches import cv2_imshow
```

## Data Processing

```python
# loading the source and reference images
anfield = cv2.imread('anfield.jpeg')
anfield = cv2.cvtColor(anfield, cv2.COLOR_BGR2RGB)

campNou = cv2.imread('campNou.jpeg')
campNou = cv2.cvtColor(campNou, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1,)
plt.title("Source")
plt.axis('off')
plt.imshow(anfield)

plt.subplot(1, 2, 2)
plt.title("Reference")
plt.axis('off')
plt.imshow(campNou)
```

## Histogram Matching

```python
# histogram matching
multi = True if anfield.shape[-1] > 1 else False
matched = exposure.match_histograms(anfield, campNou, multichannel=multi)

plt.title("Histogram Matched")
plt.axis('off')
plt.imshow(matched)
```
