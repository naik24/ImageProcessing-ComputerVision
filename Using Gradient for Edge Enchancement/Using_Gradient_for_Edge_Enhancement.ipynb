## Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Edge Enhancement

```python
# loading the image
lens = cv2.imread('lens.tif')

# defining sobel filters
grad_x = cv2.Sobel(lens, -1, 1, 0, ksize=3)
grad_y = cv2.Sobel(lens, -1, 0, 1, ksize=3)

# adding the sobel filter outputs
sobel_lens = cv2.addWeighted(grad_x, 0.5, grad_y, 0.5, 0)

# plotting the results
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(lens, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("Sobel Gradient")
plt.imshow(sobel_lens, cmap = 'gray')
```
